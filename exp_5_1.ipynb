```python
import pandas as pd

data = {
    'Weather': ['Sunny', 'Sunny', 'Rainy', 'Rainy'],
    'PlayGame': ['Yes', 'Yes', 'No', 'No']
}
df = pd.DataFrame(data)
display(df)

# 2. Calculate Prior Probabilities P(Class)
total_instances = len(df)
p_yes = len(df[df['PlayGame'] == 'Yes']) / total_instances
p_no = len(df[df['PlayGame'] == 'No']) / total_instances

print(f"P(PlayGame = Yes): {p_yes:.2f}")
print(f"P(PlayGame = No): {p_no:.2f}")

# 3. Calculate Likelihoods P(Feature | Class)

# For Class 'PlayGame = Yes'
df_yes = df[df['PlayGame'] == 'Yes']
p_sunny_given_yes = len(df_yes[df_yes['Weather'] == 'Sunny']) / len(df_yes)
p_rainy_given_yes = len(df_yes[df_yes['Weather'] == 'Rainy']) / len(df_yes)

print("\nLikelihoods for PlayGame = Yes:")
print(f"P(Weather = Sunny | PlayGame = Yes): {p_sunny_given_yes:.2f}")
print(f"P(Weather = Rainy | PlayGame = Yes): {p_rainy_given_yes:.2f}")

# For Class 'PlayGame = No'
df_no = df[df['PlayGame'] == 'No']
p_sunny_given_no = len(df_no[df_no['Weather'] == 'Sunny']) / len(df_no)
p_rainy_given_no = len(df_no[df_no['Weather'] == 'Rainy']) / len(df_no)

print("\nLikelihoods for PlayGame = No:")
print(f"P(Weather = Sunny | PlayGame = No): {p_sunny_given_no:.2f}")
print(f"P(Weather = Rainy | PlayGame = No): {p_rainy_given_no:.2f}")

# 4. Prediction for 'Weather = Sunny'

# Calculate numerator for P(PlayGame = Yes | Weather = Sunny)
numerator_yes = p_sunny_given_yes * p_yes

# Calculate numerator for P(PlayGame = No | Weather = Sunny)
numerator_no = p_sunny_given_no * p_no

print(f"\nNumerator for PlayGame = Yes (given Weather = Sunny): {numerator_yes:.2f}")
print(f"Numerator for PlayGame = No (given Weather = Sunny): {numerator_no:.2f}")

# Compare and predict
if numerator_yes > numerator_no:
    prediction = 'Yes'
else:
    prediction = 'No'

print(f"\nPrediction for Weather = Sunny: PlayGame = {prediction}")
```
