```python
import numpy as np

# Data points
points = np.array([2, 4, 5, 12])

# Number of medoids
k = 2

# Initialize medoids (first k points)
medoids = points[:k]

# Distance function (absolute distance for 1D)
def distance(a, b):
    return abs(a - b)

def total_cost(points, medoids):
    cost = 0
    for p in points:
        distances = [distance(p, m) for m in medoids]
        cost += min(distances)
    return cost

# PAM Algorithm
for iteration in range(10):
    best_medoids = medoids.copy()
    min_cost = total_cost(points, medoids)

    for i in range(len(medoids)):
        for p in points:
            if p not in medoids:
                new_medoids = medoids.copy()
                new_medoids[i] = p

                cost = total_cost(points, new_medoids)

                if cost < min_cost:
                    min_cost = cost
                    best_medoids = new_medoids

    if np.array_equal(best_medoids, medoids):
        break

    medoids = best_medoids

# Assign clusters
clusters = {m: [] for m in medoids}
for p in points:
    distances = [distance(p, m) for m in medoids]
    nearest = medoids[np.argmin(distances)]
    clusters[nearest].append(p)

# Output
print("Final Medoids:", medoids)
print("Clusters:")
for m, pts in clusters.items():
    print(f"Medoid {m}: {pts}")
```
